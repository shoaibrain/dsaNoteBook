## Valid Parentheses

open bracket, then it means its a closing bracket. So pop the item from the stack, and check if currentItem and stack popped item are valid pair, if it is a valid pair, continue else return False aka not valid parentheses
at the end of iteration the stack should be empty because all the opening and closing brackets should cutoff each other

```
def algorithm(input):
    1. Iterate over input char by char (input is assumed to be string of parentheses)
    2. If iterator value is opening braces {[(; push into stack
    2. If iterator value is closing braces }]); pop from stack
     
```
